```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats
import math
from sklearn.model_selection import train_test_split
from sklearn import metrics
from hklearn.naive_bayes import BernoulliNB, MultinomialNB
```

### Nota general: En este ejercicio se implementó Naive Bayes para las distribuciones de Bernoulli y multinomial tanto con estimadores EMV y MAP, ambas implementaciones se están importando desde el módulo hklearn.naive_bayes que se creó, para ver las implementaciones basta con ir al archivo hklearn/naive_bayes.py

## Comenzamos cargando el dataframe de los correos spam.csv

```python
df = pd.read_table('spam.csv')
```

### Los datos de cada renglón vienen en una cadena, donde cada dato de cada columna viene separado por un espacio, por lo tanto utilizamos una separación simple de la cadena en una lista y reasignamos al renglón el conjunto de valores de cada columna

```python
def toFloat(x):
    try:
        return float(x)
    except:
        print(x)
        return 0.
n = len(str(list(df.iloc[0])).split())
m = df.shape[1]
df_processed = pd.DataFrame(np.zeros((m,n)))
for i in range(df.shape[0]):
    a = str(list(df.iloc[i])).split()
    a[0] = a[0].lstrip("['")
    a[-1] = a[-1].rstrip("]'")
    li = list(map(toFloat, a))
    df_processed.at[i, :] = np.array(li)
```

```python
df_processed
```

### Guardamos los datos procesados en otro archivo csv para cargarlos posteriormente de manera directa

```python
df_processed.to_csv("spam_processed.csv")
```

```python
df = pd.read_csv("spam_processed.csv")
```

```python
df = df.drop(axis = 1, labels = ['Unnamed: 0'])
```

```python
df
```

### Porcentajes de correos spam y no spam

```python
print("Porcentaje etiquetado como spam: " + str(100. * df[df['2000'] == 1.0].shape[0]/df.shape[0]))
print("Porcentaje etiquetado como no spam: " + str(100. * df[df['2000'] == 0.0].shape[0]/df.shape[0]))
```

### Dividiendo aleatoriamente el conjunto de datos en entrenamiento, validación y prueba

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=0)
```

### Comenzaremos haciendo uso de un modelo bayesiano ingenuo, asumiendo una distribución de bernoulli tanto para las clases como para los atributos. Este modelo fue programado para esta tarea y es parte del módulo hklearn (homekitlearn) el cual se irá construyendo a lo largo de las tareas para esta clase.
#### Al utilizar Bernoulli para el caso de los atributos, se tendrá que las $n$ ocurrencias de cada palabra correspondiente a cada atributo, se reduciran a si hubo o no ocurrencia.

#### Hacemos uso del estimador máximo a posteriori con una $\alpha = 2$ para entrenar nuestro modelo BernoulliNB y hacemos las predicciones correspondientes al conjunto de entrenamiento y prueba.

```python
model = BernoulliNB(estimator='MAP', alpha = 2.0)
model.fit(X_train, y_train)
y_pred_train = model.predict(X_train) 
print(y_pred_train)
print(f"Reporte de clasificación para el clasificador casero BernoulliNB (maximo a posteriori) en los datos de entrenamiento:\n"
      f"{metrics.classification_report(y_train, y_pred_train)}\n")
y_pred_val = model.predict(X_val) 
print(y_pred_val)
print(f"Reporte de clasificación para el clasificador casero BernoulliNB (maximo a posteriori) en los datos de validación:\n"
      f"{metrics.classification_report(y_val, y_pred_val)}\n")
```

#### Hacemos uso del estimador de máxima verosimilitud para entrenar nuestro modelo BernoulliNB y hacemos las predicciones correspondientes al conjunto de entrenamiento y prueba.

```python
model2 = BernoulliNB(estimator='ML')
model2.fit(X_train, y_train)
y_pred_train = model2.predict(X_train) 
print(y_pred_train)
print(f"Reporte de clasificación para el clasificador casero BernoulliNB (maxima verosimilitud) en los datos de entrenamiento:\n"
      f"{metrics.classification_report(y_train, y_pred_train)}\n")
y_pred_val = model2.predict(X_val) 
print(y_pred_val)
print(f"Reporte de clasificación para el clasificador casero BernoulliNB (maxima verosimilitud) en los datos de validación::\n"
      f"{metrics.classification_report(y_val, y_pred_val)}\n")
```

#### Para el caso de Bernoulli con el estimador MAP, se tuvo un rendimiento mayor en términos de precisión de las predicciones con respecto al EMV, de hecho este modelo fue el que tuvo un mayor rendimiento, pese a la pérdida de información derivada de asumir cada variable como binaria.

### Para el segundo modelo vamos a asumir una distribución multinomial tanto para las clases como para los atributos. Esta distribución se asume porque cada atributo de cada ejemplo toma valores enteros que son las ocurrencias de palabras, cada ejemplo se puede ver como un vector aleatorio $\mathbf{x} = (x_1, ..., x_{K})$, donde $x_j$ es el número de ocurrencias de la palabra j-esima.

#### Hacemos uso del estimador máximo a posteriori con una $\alpha = 2$ y una $\beta$ correspondiente al número de clases para los estimadores de la clase, y al número de atributos para los estimadores de los atributos dada la clase, para entrenar nuestro modelo MultinomialNB y hacemos las predicciones correspondientes al conjunto de entrenamiento y prueba.

```python
model3 = MultinomialNB(estimator='MAP', alpha = 2.0)
model3.fit(X_train, y_train)
y_pred_train = model3.predict(X_train) 
print(y_pred_train)
print(f"Reporte de clasificación para el clasificador casero MultinomialNB (maximo a posteriori) en los datos de entrenamiento:\n"
      f"{metrics.classification_report(y_train, y_pred_train)}\n")
y_pred_val = model3.predict(X_val) 
print(y_pred_val)
print(f"Reporte de clasificación para el clasificador casero MultinomialNB (maximo a posteriori) en los datos de validación:\n"
      f"{metrics.classification_report(y_val, y_pred_val)}\n")
```

#### Hacemos uso del estimador de máxima verosimilitud para entrenar nuestro modelo MultinomialNB y hacemos las predicciones correspondientes al conjunto de entrenamiento y prueba.

```python
model4 = MultinomialNB(estimator='ML')
model4.fit(X_train, y_train)
y_pred_train = model4.predict(X_train) 
print(y_pred_train)
print(f"Reporte de clasificación para el clasificador casero MultinomialNB (maxima verosimilitud) en los datos de entrenamiento:\n"
      f"{metrics.classification_report(y_train, y_pred_train)}\n")
y_pred_val = model4.predict(X_val) 
print(y_pred_val)
print(f"Reporte de clasificación para el clasificador casero MultinomialNB (maxima verosimilitud) en los datos de validación:\n"
      f"{metrics.classification_report(y_val, y_pred_val)}\n")
```
